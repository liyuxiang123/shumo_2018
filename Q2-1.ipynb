```python
import numpy as np
import pandas as pd
pi = 3.141592653

city = ['哈尔滨','北京','西安','郑州','成都','重庆','武汉','上海','昆明','广州','乌鲁木齐','拉萨']
J = [126.54,116.40,108.94,113.62,104.06,106.55,114.31,121.47,102.83,113.26,87.62,91.11]
W = [45.80,39.90,34.34,34.75,30.57,29.57,30.59,31.23,24.88,23.13,43.83,29.64]
P = [10.93,37.3,9.61,9.56,16,30.5,10.91,24.2,6.7,27,2.2,0.9]

V = [[0 for i in range(12)] for i in range(12)]
```

```python
for i in range(0,12,1):
    for j in range(0,12,1):
        if i != j:
            temp = 6371* np.sqrt(np.sum(np.square(J[i] - J[j])+np.square(W[i]-W[j]))) * pi / 180
            if temp < 3000:
                V[i][j] = 8
            if temp < 1200:
                V[i][j] = 16
            if temp < 600:
                V[i][j] = 32
print V
```

```python
V1 = pd.DataFrame(V,columns = city,index = city)
print V1
```

```python
V2 = pd.DataFrame(V,columns = city,index = city)
print V2
```
